# Supermarket analysis

In this file, I'll analyze a supermarket sales dataset. First, I'll check if it is clean and contains relevant data. Next, I'll be diving into the numbers and statistics to understand the relation between variables.

```python
# Importing data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import datetime as dt

dados = pd.read_csv('C:\\Users\\Sergio\\Desktop\\Danilo\\Estudos\\Python\\supermarket_sales - Sheet1.csv')
dados = pd.DataFrame(dados)
dados.head(50)
```

```python
# Checking data (select distinct)

print('City\n', dados['City'].unique(), '\n')
print('Product line\n', dados['Product line'].unique(), '\n')
print('MaxUnit price\n', dados['Unit price'].max(), '\n')
print('MinUnit price\n', dados['Unit price'].min(), '\n')
print('Payment\n', dados['Payment'].unique(), '\n')
```

```python
dados.info()
```

```python
# Main statistics

dados.describe()
```

```python
# Grouping data by gender and formatting to 2 decimal places

dados.groupby(['Gender']).agg({'Unit price': ['count', 'min', 'mean', 'max'], 'Total': ['count', 'min', 'mean', 'max']}).style.format('{:.2f}')
```

The sample has no significant difference in terms of female and male shop values

```python
# Other groups

tabela = dados.groupby('City').agg({'Invoice ID': ['count'], 'Total': ['mean']})

tabela.style.format({('Total', 'mean'): "{:.2f}"})
```

Number of sales and total values are quite similar in all cities

```python
dados.groupby(['City']).agg({'Rating': ['count', 'min', 'mean', 'max'],}).style.format('{:.2f}')
```

Mandalay has a lower average rating comparing to Naypyitaw and Yangon

```python
columns = ['Total', 'Payment']
payment_method = dados[columns].copy()

payment_method.groupby(['Payment']).describe()
```

The sample shows that people buy more on cash or Ewallet than Credit card, but the total value spent in each payment method is similar

### Charts

**Pie chart that shows the percentage that each product line represents in total sales**

```python
# Group by product line
productline = dados.groupby('Product line').sum()[['Total']]

# Calculating percentages
productline['percentage'] = productline['Total'] / productline['Total'].sum() * 100

# Creating chart
plt.pie(
    productline['percentage'], # values
    labels = productline.index, # names
    autopct = '%1.1f%%', # format to 1 decimal place
    wedgeprops = {'linewidth': 3, 'edgecolor' : 'white'}) # Details

plt.show()
```

The total sales value is well distributed between product categories

**Bar chart grouped by city that shows the average total sales for each month**

```python
dados['Date'] = pd.to_datetime(dados['Date']) # converting 'Date' column from string to date
dados['month'] = dados['Date'].dt.strftime('%Y-%m') # creating 'Month' column from 'Date'

# Pivot table that groups data by city and month
dados_porms = dados.pivot_table(index='month', columns='City', values='Total', aggfunc='sum', fill_value=0) 

# Printing data that should be exhibited on the chart
dados_porms = (dados_porms/1000) # divide os valores por 1000
dados_porms = pd.DataFrame(dados_porms)
print(dados_porms)

# Creating list of values for each city
mandalay = dados_porms.iloc[0].tolist()
naypyitaw = dados_porms.iloc[1].tolist()
yangon = dados_porms.iloc[2].tolist()

# Creating list of months
meses = dados_porms.columns.tolist()

plot_dados = pd.DataFrame(dados_porms, index = meses)

# Creating chart
ax = dados_porms.plot(kind = 'bar')
for p in ax.patches: # Add label for each bar
    ax.annotate(f"{p.get_height():.2f}", 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', 
                va='center', 
                xytext=(0, 5), 
                textcoords='offset points', 
                fontsize = 8)
plt.ylabel('Total (k)')
plt.ylim(0, 55)
```

Observing only the first months of the year, in terms of total sales value there's not much difference between cities. It seems that sales performance tend to go up or down regardless of the market's location
