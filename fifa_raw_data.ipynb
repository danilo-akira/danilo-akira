## Cleaning raw data
In this file, I'll be cleaning a raw dataset that contains data about the videogame Fifa 21. Each line of the data set represents a football player that exists in the game and each column represents a different atribute or information about the player

```python
# Importing libraries and reading .csv file

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import datetime as dt

dados = pd.read_csv('C:\\Users\\Sergio\\Desktop\\Danilo\\Estudos\\Python\\fifa21 raw data v2.csv', encoding = 'utf-8')
dados = pd.DataFrame(dados)
dados = dados.drop(columns = ['photoUrl', 'playerUrl'])

pd.set_option('display.max_columns', None)
dados.head(10)
```

```python
# Checking columns and data types

dados.info()
```

#### Columns to be corrected:
1. Club: remove '\n\n\n\n'
1. Height: convert data from centimeters and feet/inches to meters
1. Weight: convert data from pounds to kilos
1. Joined: convert from string to date
1. Value: remove '€' symbol and 'M' and 'K' (millions and thousands). Convert data from string to float
1. Wage: remove '€' symbol and 'K' (thousands). Convert data from string to float
1. Release clause: remove '€' symbol and 'M' and 'K' (millions and thousands). Convert data from string to float

```python
# Club

dados['Club'] = dados['Club'].str.replace('\n\n\n\n', '')
```

```python
#Height

def hconversion (string):
    if "'" in string:
        string = string.replace("\\", "").replace('"', "")
        feet, inches = map(int, string.split("'"))
        total_inches = feet * 12 + inches
        meters = total_inches * 0.0254
        return round(meters, 2)
    elif 'cm' in string:
        return float(string.replace('cm', '')) / 100
    else:
        return string

dados['Height'] = dados['Height'].apply(hconversion)

print(dados['Height'].dtype)
print(dados['Height'].unique())
```

```python
# Weight

dados['Weight'].unique()

def wconversion (string):
    if "kg" in string:
        return int(string.replace('kg', ''))
    elif 'lbs' in string:
        return round(int(string.replace('lbs', '')) / 2.205, 0)
    else:
        return string

dados['Weight'] = dados['Weight'].apply(wconversion)

print(dados['Weight'].dtype)
print(dados['Weight'].unique())
```

```python
# Joined

dados['Joined'] = pd.to_datetime(dados['Joined'])
print(dados['Joined'])
```

```python
# Value 

dados['Value'].unique()

def value(string):
    if 'M' in string:
        string = string.replace('€', '').replace('M', '')
        return float(string)
    elif 'K' in string:
        string = string.replace('€', '').replace('K', '')
        return round(float(string) / 1000, 3)
    else:
        return string

dados['Value'] = dados['Value'].apply(value)

print(dados['Value'])
```

```python
# Wage

dados['Wage'].unique()

def value(string):
    if 'K' in string:
        string = string.replace('€', '').replace('K', '')
        return float(string)
    else:
        string = string.replace('€', '')
        return round(float(string) / 1000, 3)
    
dados['Wage'] = dados['Wage'].apply(value)

print(dados['Wage'])
```

```python
# Release Clause

dados['Release Clause'].unique()

def release_clause(string):
    if 'M' in string:
        string = string.replace('€', '').replace('M', '')
        return float(string)
    elif 'K' in string:
        string = string.replace('€', '').replace('K', '')
        return round(float(string) / 1000, 3)
    else:
        return string

dados['Release Clause'] = dados['Release Clause'].apply(release_clause)

print(dados['Release Clause'])
```

```python
# Changing column names

dados_clean = dados.rename(columns={
    'Height': 'Height (m)',
    'Weight': 'Weight (kg)',
    'Value': 'Value (MM)',
    'Wage': 'Wage (k)',
    'Release Clause': 'Release Clause (MM)'
}, inplace = False)

dados_clean.head(10)
```

```python
# Download clean table in .csv format

dados_clean.to_csv('dados_fifa.csv', index = False, encoding = 'utf-8')
```
